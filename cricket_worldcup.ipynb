```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import json
```

```python

def get_data_frame_from_leauge_match(leauge_match):
    json_data_arr=[]
    for file in os.listdir(leauge_match):
        json_match_file_name = leauge_match+'/'+file
        if json_match_file_name.__contains__('.json'):
            json_match_file=json.load(open(json_match_file_name))
            print(json_match_file_name)
            city = json_match_file['info']['city'] if 'city' in json_match_file['info'] else None
            # city = json_match_file['info']['city']
            dates = json_match_file['info']['dates'][0]
            is_match_has_result = 'result' if 'result' in json_match_file['info']['outcome'] else 'winner'
            winner = json_match_file['info']['outcome'][is_match_has_result]
            run_or_wikets = None
            by_run_by_wicket=None
            if is_match_has_result == 'winner':
                if 'method' in json_match_file['info']['outcome']:
                    winner_out='method'
                    run_or_wikets=json_match_file['info']['outcome'][winner_out]
                elif 'runs' in json_match_file['info']['outcome']['by']:
                    winner_out='runs'
                    run_or_wikets=json_match_file['info']['outcome']['by'][winner_out]

                else:
                    winner_out='wickets'
                    run_or_wikets=json_match_file['info']['outcome']['by'][winner_out]
                
                by_run_by_wicket = winner_out
            
            # print(json_match_file_name)
            if len(json_match_file['innings'])==2:
                team_1st_inning = json_match_file['innings'][0]['team']
                team_2nd_inning = json_match_file['innings'][1]['team']
            else:
                team_1st_inning = json_match_file['innings'][0]['team']
                team_2nd_inning = None
            # json_data = {'city':city, 'dates':dates, 'winner':winner,'run_or_wikets':run_or_wikets,'by_run_by_wicket':by_run_by_wicket,
            #              'team_1st_inning':team_1st_inning, 'team_2nd_inning':team_2nd_inning,
            #              }
            toss_winner = json_match_file['info']['toss']['winner']
            toss_decision = json_match_file['info']['toss']['decision']
            venue = json_match_file['info']['venue']
            
            for innigs in range(len(json_match_file['innings'])):
                overs = json_match_file['innings'][innigs]['overs']
                # print(len(overs))
                for over in overs:
                    over_number = over['over']
                    for ball in over['deliveries']:
                        batter = ball['batter']
                        bowler = ball['bowler']
                        non_striker = ball['non_striker']
                        runs_batter = ball['runs']['batter']
                        runs_extras = ball['runs']['extras']
                        runs_total = ball['runs']['total']
                        json_data = {'match_number':file.split('.')[0],'city':city, 'venue':venue, 'dates':dates, 'winner':winner,'run_or_wikets':run_or_wikets,'by_run_by_wicket':by_run_by_wicket,
                        'team_1st_inning':team_1st_inning, 'team_2nd_inning':team_2nd_inning,'over_number':over_number,'toss_winner':toss_winner,'toss_decision':toss_decision,
                        'batter':batter,'bowler':bowler,'non_striker':non_striker,'runs_batter':runs_batter, 'runs_extras':runs_extras, 'runs_total':runs_total
                        }
                        json_data_arr.append(json_data)
        
    return pd.DataFrame(json_data_arr)
    
```

```python
df = pd.DataFrame()
match_number=1
for dataset in os.listdir('dataset'):
    df = pd.concat([df, get_data_frame_from_leauge_match('dataset/'+dataset)])
    
```
